Plot data generated by Flow*, in Julia.

To install Flow* i've written the following instructions: https://gist.github.com/mforets/86958a476b4f9ffe1884171d0e8966aa

```julia
using DelimitedFiles, LazySets, Plots
```

```julia
data = readdlm("ltv_test.plt", skipstart=10);  # con skipstart descarto los datos de configuracion de gnuplot
data = data[1:end-1, :]; # la ultima linea tiene un "e" de "end"
data = convert(Matrix{Float64}, data);
NVERTICES = 10 # number of points per polygon... is it constant?
NPOLYS = Int(size(data, 1) / NVERTICES)
```

```julia
array = [VPolygon(data[1+NVERTICES*(i-1):NVERTICES*i, :]') for i in 1:NPOLYS];
```

```julia
plot(array)
```
